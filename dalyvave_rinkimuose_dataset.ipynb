```python
import pandas as pd

DATA_FILE_URL = "https://atviriduomenys.vrk.lt/datasets/gov/vrk/Balsave/:format/csv"

BALSAI = "Balsave.csv"


def read_data() -> pd.DataFrame:
    data = pd.read_csv(BALSAI)
    data.drop(index=data.index[-1], axis=0, inplace=True)
    data["balsavusiu_dalis"] = data["atvyko_balsuoti"]/data["rinkeju_sarase"]
    return data

display(read_data())
```

```python
df = read_data()

columns = ['rink_turo_id', 'rink_turo_pavad']
rink_turai = df[columns].drop_duplicates().sort_values('rink_turo_id', ascending=True)

display(rink_turai)
```

```python
df = df[(df['sav_pavadinimas'] == 'Kauno miesto savivaldybė') & (df['rink_turo_id'] == 1744)]

columns = ['apyl_id', 'apyl_pavad','rinkeju_sarase']

apylinkes = df.groupby(['apyl_id', 'apyl_pavad'])['rinkeju_sarase'].sum().sort_values(ascending=False)

display(apylinkes)
```

```python
columns = ['amziaus_grupe']
amziaus_grupes = df[columns].drop_duplicates().sort_values('amziaus_grupe', ascending=True)

display(amziaus_grupes)
```
